```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
batch = 0

f_lengths = [0.1, 0.5, 1.4]
data = np.load(f'../data/{batch}.npy')
gt = np.load(f'../data/{batch}_y.npy')

rows, cols = 4, 8
space = 15
fig, axes = plt.subplots(rows, cols, figsize=(space * cols/rows, space))

for j in range(cols):
    for i in range(rows-1):
        axes[i, j].imshow(data[j, i], cmap='gray')
        axes[i, j].set_title(f'Focal length {f_lengths[i]}')
    axes[rows-1, j].imshow(gt[j], cmap='gray')
    axes[rows-1, j].set_title(f'GT')
plt.show()
```
